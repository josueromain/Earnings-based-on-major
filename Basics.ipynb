```python
def clean_col (col):
    col = col.replace("ShareWomen","share_women")
    col = col.lower()
    col = col.strip()
    return col
```

```python
new_columns = []
recent_grads.columns
for row in recent_grads.columns:
    clean_c = clean_col(row)
    new_columns.append(clean_c)
    
recent_grads.columns = new_columns 
```
